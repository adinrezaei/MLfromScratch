```python
import numpy as np
import sklearn
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
```

```python
# Load the diabetes dataset
X, t = datasets.load_diabetes(return_X_y=True)
t = t.reshape(-1,1)
```

```python
# Finding W in closed form solution
Xp = np.linalg.pinv(X)
W = np.matmul(Xp,t)
#W = np.linalg.inv(X.T*X)*(X.T)*t
```

```python
Y = np.matmul(X,W)
```

```python
# Plot the figure
Fnt = 14
plt.scatter(t,Y)
plt.xlabel('Target', fontsize = Fnt)
plt.ylabel('Prediction', fontsize = Fnt)
plt.title('Linear regression', fontsize = Fnt)
plt.xticks(fontsize = Fnt)
plt.yticks(fontsize = Fnt)
```
